```python
# Edward L. Rivera Python Code 
# The purpose of this script is to impose a W-shape structure onto the W maze for every recorded session. The script also generates the coordinates of the 6 nodes used to define the W shaped structure. This data will be use to downstream linearization of spatial tracking data obtained from DLC model  
# 
# 
# Import relevant packages 
import os
import os.path
import cv2
import numpy as np
import tkinter as tk
from tkinter import simpledialog
from tkinter import *
from numpy import savetxt
from sympy import Point, Line, pi
import math

```

```python
nodes1=[]
nodes2=[]
preview = None
lastNode = (-1, -1)
# function to display the coordinates of the points clicked on the image
def click_event(event, x, y, flags, params):
    global lastNode,img, preview
   
   # checking for left mouse clicks
    if event == cv2.EVENT_LBUTTONDOWN:
        print('Left Click')
        print(f'({x},{y})')
        
        if not nodes1:
            lastNode=(x,y)
        
        
            if len(nodes1)<6:
        
                nodes1.append(lastNode)
           
            else:
                nodes2.append(lastNode)
                
             # put coordinates as text on the image
            cv2.putText(img, f'({lastNode[0]},{lastNode[1]})', (lastNode[0], lastNode[1]),   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.circle(img, (lastNode[0], lastNode[1]), 10, (0, 0, 255), -1)
            
            # new initial point and preview is now a copy of the original image

            preview = img.copy()
            # this will be a point at this point in time
            cv2.line(preview, lastNode, (x,y), (0,255,0), 1)
                
        elif len(nodes1)==4:
            
            lastNode=nodes1[-2]
           # put coordinates as text on the image
            cv2.putText(img, f'({lastNode[0]},{lastNode[1]})', (lastNode[0], lastNode[1]),   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.circle(img, (lastNode[0], lastNode[1]), 10, (0, 0, 255), -1)
            
            # new initial point and preview is now a copy of the original image

            preview = img.copy()
            # this will be a point at this point in time
            cv2.line(preview, lastNode, (x,y), (0,255,0), 1)
            
        elif len(nodes2)==4:
            lastNode=nodes2[-2]
           # put coordinates as text on the image
            cv2.putText(img, f'({lastNode[0]},{lastNode[1]})', (lastNode[0], lastNode[1]),   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.circle(img, (lastNode[0], lastNode[1]), 10, (0, 0, 255), -1)
            
            # new initial point and preview is now a copy of the original image

            preview = img.copy()
            # this will be a point at this point in time
            cv2.line(preview, lastNode, (x,y), (0,255,0), 1)
            
        elif len(nodes1)==6 and not nodes2:
            lastNode=(x,y)
            if len(nodes2)<6:
        
                nodes2.append(lastNode)
           
            
                
             # put coordinates as text on the image
            cv2.putText(img, f'({lastNode[0]},{lastNode[1]})', (lastNode[0], lastNode[1]),   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.circle(img, (lastNode[0], lastNode[1]), 10, (0, 0, 255), -1)
            
            # new initial point and preview is now a copy of the original image

            preview = img.copy()
            # this will be a point at this point in time
            cv2.line(preview, lastNode, (x,y), (0,255,0), 1)
            
        else:
            
            
            # put coordinates as text on the image
            cv2.putText(img, f'({lastNode[0]},{lastNode[1]})', (lastNode[0], lastNode[1]),   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.circle(img, (lastNode[0], lastNode[1]), 10, (0, 0, 255), -1)
            
            # new initial point and preview is now a copy of the original image

            preview = img.copy()
            # this will be a point at this point in time
            cv2.line(preview, lastNode, (x,y), (0,255,0), 1)
    
    elif event == cv2.EVENT_MOUSEMOVE:
        if preview is not None:
            # copy the original image again a redraw the new line
            preview = img.copy()
            cv2.line(preview, lastNode, (x,y), (0,255,0), 1)
            distance=((lastNode[0] - x)**2 + (lastNode[1] - y)**2)**0.5
            distance=round(distance,1)
            cv2.putText(preview, 'Distance='+ f'({distance})', (700, 500),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
            # using Line() method
            
            if len(nodes1)>1 and len(nodes1)<6:
                if len(nodes1)==3:
                    l1 = Line(nodes1[-2], nodes1[-1])
                    l2 = Line(nodes1[-1], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                    
                if len(nodes1)==4:
                    l1 = Line(nodes1[-2], nodes1[-1])
                    l2 = Line(nodes1[-2], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                    
                elif len(nodes1)==5:
                
                    l1 = Line(nodes1[-3], nodes1[-1])
                    l2 = Line(nodes1[-1], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                    
                else:
                    l1 = Line(nodes1[-2], nodes1[-1])
                    l2 = Line(nodes1[-1], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                
            elif len(nodes2)>1 and len(nodes2)<6:
                if len(nodes2)==3:
                    l1 = Line(nodes2[-2], nodes2[-1])
                    l2 = Line(nodes2[-1], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                    
                elif len(nodes2)==4:
                    l1 = Line(nodes2[-2], nodes2[-1])
                    l2 = Line(nodes2[-2], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                    
                elif len(nodes2)==5:
                
                    l1 = Line(nodes2[-3], nodes2[-1])
                    l2 = Line(nodes2[-1], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                else:
                    l1 = Line(nodes2[-2], nodes2[-1])
                    l2 = Line(nodes2[-1], (x, y))
                    angle = l1.angle_between(l2)
                    angle=math.degrees(angle)
                    angle=round(angle,1)
                    cv2.putText(preview, 'Angle='+ f'({angle})', (700, 450),   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                        
            
            
            
    elif event == cv2.EVENT_LBUTTONUP:
        # if we are drawing, preview is not None and since we finish, draw the final line in the image
        if preview is not None:
            preview = None
            cv2.line(img, lastNode, (x,y), (255,0,0), 1)
            cv2.putText(img, f'({x},{y})', (x, y),   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.circle(img, (x, y), 10, (0, 0, 255), -1)
            #cv2.putText(img, f'({distance})', ((x-lastNode[0])/2, y), 
            lastNode=(x,y)
            
            if len(nodes1)<6:
        
                nodes1.append(lastNode)
           
            else:
                nodes2.append(lastNode)
           
```

```python
cohort='Cohort-6'
dataFolder='H:/Social W Behavioral Data/'+cohort
os.chdir(dataFolder)
os.listdir()

path = dataFolder
for root, dirs, files in os.walk(path):
    for _file in files:
           #print(_file)
        if _file.endswith(".mp4"):
            sessionVideo=root+'/'+_file
            
            print(sessionVideo)
            
            if os.path.isfile(sessionVideo.replace(".mp4",".jpeg"))==False:
                vidcap = cv2.VideoCapture(sessionVideo)
                success,image = vidcap.read(1)
                cv2.imwrite(sessionVideo.replace(".mp4",".jpeg"), image)     # save frame as JPEG file    
            frame=sessionVideo.replace(".mp4",".jpeg")  
                
            frame_rat1=frame.replace(".1.jpeg","-Rat1.npy")
            frame_rat2=frame.replace(".1.jpeg","-Rat2.npy")
                
            if os.path.isfile(frame_rat1)==False:
                nodes1=[]
                nodes2=[]
                preview = None
                lastNode = (-1, -1)


   
  
                # set the named window and callback    
                img = cv2.imread(frame)
                cv2.namedWindow(frame)
                cv2.setMouseCallback(frame, click_event)

                while (True):
                # if we are drawing show preview, otherwise the image
                    if preview is None:
                        cv2.imshow(frame,img)
                    else:
                        cv2.imshow(frame,preview)
                    k = cv2.waitKey(1) & 0xFF
                    if k == ord('q'):
                        break;
        
                    elif len(nodes2)==6:
                        break

                cv2.destroyAllWindows()
                cv2.waitKey(0)
                
                np.save(frame_rat1,nodes1)
                np.save(frame_rat2,nodes2)
                

           
          
```
